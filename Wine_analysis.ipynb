```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  
%matplotlib inline
```

```python
data = pd.read_csv('Wine_Climate_Data.csv')
```

```python
data.head()
```

```python
#average rainfall/average score
import matplotlib.pyplot as plt
data.plot(x='Avg Rainfall (Inch)', y='Avg Score', style='o')  
plt.title('Average Rainfall to Average Score correlation')  
plt.xlabel('Average Rainfall')  
plt.ylabel('Average Score')
plt.axis([20, 75, 80, 92.5])
plt.savefig('Average Rainfall to Average Score correlation.png')
plt.show()
```

```python
#average rainfall/average value for money score
data.plot(x='Avg Rainfall (Inch)', y='Avg Value For Money Score', style='o')  
plt.title('Average Rainfall to Average Value For Money Score correlation')  
plt.xlabel('Average Rainfall')  
plt.ylabel('Average Value For Money Score')
plt.axis([20, 75, 88, 95])
plt.savefig('Average Rainfall to Average Value For Money Score correlation')
plt.show()  
```

```python
#average rainfall/Avg Price ($)
data.plot(x='Avg Rainfall (Inch)', y='Avg Price ($)', style='o')  
plt.title('Average Rainfall to Average price correlation')  
plt.xlabel('Average Rainfall')  
plt.ylabel('Average Price')
plt.axis([20, 75, 10, 40])
plt.savefig('Average Rainfall to Average price correlation.png')
plt.show()
```

```python
#average humidity/Avg Score
data.plot(x='Avg Humidity', y='Avg Score', style='o')  
plt.title('Average Humidity to Averoge Score correlation')  
plt.xlabel('Average Humidity')  
plt.ylabel('Average Score')
plt.axis([45, 90, 80, 92.5])
plt.savefig('Average Humidity to Averoge Score correlation.png')
plt.show()  
```

```python
#average humidity/Avg Value For Money Score
data.plot(x='Avg Humidity', y='Avg Value For Money Score', style='o')  
plt.title('Average Humidity to Average Value For Money Score correlation')  
plt.xlabel('Average Humidity')  
plt.ylabel('Average Value For Money Score')
plt.axis([45, 90, 88, 95])
plt.savefig('Average Humidity to Average Value For Money Score correlation')
plt.show()  
```

```python
#average humidity/Avg Price ($)
data.plot(x='Avg Humidity', y='Avg Price ($)', style='o')  
plt.title('Average Humidity to Average price correlation')  
plt.xlabel('Average Humidity')  
plt.ylabel('Average Price')
plt.axis([45, 90, 10, 40])
plt.savefig('Average Humidity to Average price correlation.png')
plt.show()  
```
